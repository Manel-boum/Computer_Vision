```python
# Anonymize all faces in a VIDEO

# Install required libraries (only if not already installed)
!pip install opencv-python mediapipe  

import os
import cv2
import mediapipe as mp

# Function to process a frame and anonymize detected faces
def process_img(img, face_detection, H, W):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    out = face_detection.process(img_rgb)

    if out.detections is not None:
        for detection in out.detections:
            location_data = detection.location_data
            bbox = location_data.relative_bounding_box

            x1, y1, w, h = bbox.xmin, bbox.ymin, bbox.width, bbox.height

            # Calculate absolute pixel values for bounding box
            x1 = int(x1 * W)
            y1 = int(y1 * H)
            w = int(w * W)
            h = int(h * H)

            # Blur only the face region
            img[y1:y1 + h, x1:x1 + w, :] = cv2.blur(img[y1:y1 + h, x1:x1 + w, :], (50, 50))

    return img

# Simulate command-line arguments with a dictionary
args = {
    "mode": "video",  # Mode selection
    "filePath": "./data/testVideo.mp4"  # Path to the input video
}

# Create output directory if it doesn't exist
output_dir = './output'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Initialize MediaPipe face detection
mp_face_detection = mp.solutions.face_detection

with mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.5) as face_detection:
    if args["mode"] == "video":
        # Capture the video
        cap = cv2.VideoCapture(args["filePath"])
        
        # Check if video opened successfully
        if not cap.isOpened():
            print("Error: Could not open video.")
            exit()

        # Get video properties
        frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        fps = int(cap.get(cv2.CAP_PROP_FPS))

        # Prepare the output video writer
        output_video_path = os.path.join(output_dir, 'output.mp4')
        output_video = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

        # Process each frame in the video
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break  # Exit loop if no more frames

            # Process the frame to anonymize faces
            processed_frame = process_img(frame, face_detection, frame_height, frame_width)
            
            # Write the processed frame to the output video
            output_video.write(processed_frame)

        # Release resources
        cap.release()
        output_video.release()
        print(f"Video saved to {output_video_path}")
```

```python
# Anonymize the Web Cam of our laptop 

# Install required libraries (only if not already installed)
!pip install opencv-python mediapipe  

import os
import cv2
import mediapipe as mp

# Function to process a frame and anonymize detected faces
def process_img(img, face_detection):
    H, W, _ = img.shape
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    out = face_detection.process(img_rgb)

    if out.detections is not None:
        for detection in out.detections:
            location_data = detection.location_data
            bbox = location_data.relative_bounding_box

            x1, y1, w, h = bbox.xmin, bbox.ymin, bbox.width, bbox.height

            # Calculate absolute pixel values for bounding box
            x1 = int(x1 * W)
            y1 = int(y1 * H)
            w = int(w * W)
            h = int(h * H)

            # Blur only the face region
            img[y1:y1 + h, x1:x1 + w, :] = cv2.blur(img[y1:y1 + h, x1:x1 + w, :], (50, 50))

    return img

# Simulate command-line arguments with a dictionary
args = {
    "mode": "webcam",  # Mode selection for webcam usage
}

# Initialize MediaPipe face detection
mp_face_detection = mp.solutions.face_detection

with mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.5) as face_detection:
    if args["mode"] == "webcam": 
        # Start webcam capture
        cap = cv2.VideoCapture(0)

        if not cap.isOpened():
            print("Error: Could not access the webcam.")
            exit()

        while True:
            ret, frame = cap.read()
            if not ret:
                print("Error: Could not read frame from webcam.")
                break

            # Process the frame to anonymize faces
            processed_frame = process_img(frame, face_detection)

            # Display the frame with blurred faces
            cv2.imshow('Anonymized Webcam Feed', processed_frame)

            # Break the loop if 'q' is pressed
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        # Release resources
        cap.release()
        cv2.destroyAllWindows()
```
